**Step 1: Plotting using Seaborn**

Find the regression plot using Seaborn.
Mean absolute error, mean squared error, mean absolute percentage error

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Advertising.csv')
data
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

sns.histplot(data['TV'], kde=True, ax=axes[0], color='blue')
axes[0].set_title('TV Advertising Distribution')

sns.histplot(data['Radio'], kde=True, ax=axes[1], color='green')
axes[1].set_title('Radio Advertising Distribution')

sns.histplot(data['Newspaper'], kde=True, ax=axes[2], color='orange')
axes[2].set_title('Newspaper Advertising Distribution')

plt.tight_layout()
# plt.legend()
plt.show()
```

```python

#Pair plot to explore relationships among TV, Radio, Newspaper, and Sales
# sns.pairplot(data, vars=['TV', 'Radio', 'Newspaper', 'Sales'])
sns.pairplot(data, vars=['TV', 'Radio', 'Newspaper'])

plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.tree import DecisionTreeRegressor

df = data.copy()
x = df.iloc[:,:-1].values
y = df.iloc[:,-1].values

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=42)

model = DecisionTreeRegressor()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
print("R-squared (RÂ²):", r2)

#Calculating the mean aboslute percentage error (MAPE)

```
